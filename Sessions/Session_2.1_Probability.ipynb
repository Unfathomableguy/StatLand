```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# plotting styles
plt.style.use("ggplot")
```

## Random Ball Pick Game 
### Case(1)

```python
# Blue balls
blue_balls = 998
red_balls = 2
```

```python
# random numbers generator
gen = np.random.RandomState(seed=100)
# Members Played Game Data
df_case_1 = pd.DataFrame({"Anto":gen.choice([1,0], size = 100, replace = True, p=[.998, .002]),
                          "Chandini":gen.choice([1,0], size = 100, replace = True, p=[.998, .002]),
                          "Manoj":gen.choice([1,0], size = 100, replace = True, p=[.998, .002]),
                          "Ramesh":gen.choice([1,0], size = 100, replace = True, p=[.998, .002]),
                          "Shiv":gen.choice([1,0], size = 100, replace = True, p=[.998, .002])})
df_case_1.head()
```

```python
df_1 = pd.DataFrame(df_case_1.sum(), columns = ["# Total Won"])
df_1["Ttl_Playing_Cost"] = 100*100000
df_1["Net Profit"] = df_1["# Total Won"]*1000000 - df_1["Ttl_Playing_Cost"]
df_1["Ttl_Playing_Cost"] = df_1["Ttl_Playing_Cost"]/1000000
df_1["Net Profit"] = df_1["Net Profit"]/1000000
df_1.rename(columns = {"Ttl_Playing_Cost":"Ttl_Playing_Cost(in Millions)","Net Profit":"Net Profit(in Millions)"}, inplace=True)
df_1
```
