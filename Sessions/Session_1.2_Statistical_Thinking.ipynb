Example : Assume we have `Rs 100,000` that accrues a varying rate of interest each year for `5 years`:  
> `Annual interest rates: 1%, 9%, 6%, 2%, 15%`

*Take a shortcut to find our average annual interest rate, & thus our total amount of money after 5 years, so we try to “average” these rates:*
> `(.01 + .09 + .06 + .02 + .15) ÷ 5 = .066 that is  6.6%`

*Calculate the total return after `5 years` of investment*
$$A=P{(1+\frac{r}{n})^{nt}}{\textit{ , where the symbols have their usual meanings. }}$$

```python
def compound_interst():
    """
    Calculate the compound interest
    """
    principal = int(input("The amount to be invested : "))
    rate=float(input("Interst rate : "))
    time=int(input("Time period of investment : "))
    num_times_int_applied = int(input("Number of times interest applied per time period : "))
    total_return = principal*((1+rate/100)**time)
    return print(f"The total return after {time} years is Rs {round(total_return,2)}")

# calculate the return
compound_interst()
```

* *Let’s do this the long way & compare results:*
> `Year 1: 100,000 + (100,000 * .01) = 100,000 * 1.01 = Rs 101,000`   
> `Year 2: 101,000 * 1.09 = RS 110,090`   
> `Year 3: 110,090 * 1.06 = Rs 116,695.40`  
> `Year 4: 116,695.40 * 1.02 = Rs 119,029.31`  
> `Year 5: 119,029.31 * 1.15 = Rs 136,883.70`  
> `Actual final total = Rs 136,883.70`

$\color{red}{\textit{Our shortcut overestimated our actual earnings by nearly Rs 770. We applied an additive operation to a multiplicative process, and got an inaccurate result.}}$

**Now let’s try again with the geometric mean:**
> `1.01 * 1.09 * 1.06 * 1.02 * 1.15 = 1.368837042`  
> `5th root of 1.368837042 = 1.064805657`  
>`Geometric mean = 1.064805657`  
> `Interst rate = 1.064805657-1 = 0.0648056 i.e. 6.48%`

```python
# calculate the return
compound_interst()
```

```python
# Example

## Movie A
imdb_A = 4.5 # considering out of 5*20
rotten_tomatoes_A = 68

## Movie B
imdb_B = 3
rotten_tomatoes_B = 75
```

```python
# Arithemetic Mean
am_Movie_A = (4.5+68)/2
am_Movie_B = (3+75)/2

# print
print(f"The Arithemetic Mean Of Movie_A = {am_Movie_A}\n",
      f"The Arithemetic Mean of Movie_B = {am_Movie_B}")
```

```python
# Geometric Mean
gm_Movie_A = (4.5*68)**(1/2)
gm_Movie_B = (3*75)**(1/2)

# print
print(f"The Geometric Mean Of Movie_A = {round(gm_Movie_A, 2)}\n",
      f"The Geometric Mean of Movie_B = {gm_Movie_B}")
```
