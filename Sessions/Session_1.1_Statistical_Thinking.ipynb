```python
#@title Why Median is required?
# Mean is affected by extremities!
gen = np.random.RandomState(seed=1)
random_data = pd.DataFrame({"Wages":gen.randn(20000)*5000 + 25000})
random_data_w_out = random_data.copy(deep=True)
random_data_w_out.loc[len(random_data_w_out.index)] = [500000]

# subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,6))
axes = axes.flatten()
data = [random_data,random_data_w_out]
title = ["Common Man", "Including Multibillionaries"]
for i in range(2):
    sns.histplot(data = data[i], x = "Wages", kde=True, ax = axes[i], color=["blue", "red"][i], alpha=0.2)
    axes[i].set_title(title[i])
plt.suptitle("Mean is affected by extremities!", x=0.5, y=1.09, fontsize=20, alpha=0.6)
plt.tight_layout(rect=[0, 0, 1, 1])
plt.show()
```
